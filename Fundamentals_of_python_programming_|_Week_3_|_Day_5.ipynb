Lambda Functions in Python

A lambda function is a small, anonymous (nameless) function that can have any number of parameters but only one expression.

It is useful when you need a simple function for a short period of time.

Syntax:

lambda parameters: expression

lambda → Keyword used to create the function.

Parameters → Inputs to the function.

Expression → The calculation or operation to perform.

The result of this expression is automatically returned.

```python
#Lambda Function
square = lambda x: x * x

print(square(5))
```

```python
add = lambda a, b: a + b

print(add(10, 20))
```

```python
#Why Use Lambda Functions?

#Lambda functions are useful when you need a function only once and don't want to write a full def function.

#For example, sorting a list:
```

```python
students = [("Ali", 80), ("Sara", 95), ("Ahmed", 70)]

students.sort(key=lambda student: student[1])

print(students)
```

map(), filter(), reduce() in python:

map(), filter(), and reduce() are higher-order functions in Python. They allow you to process collections (such as lists) efficiently, often with the help of lambda functions.

```python
#map()
#Purpose
#map() applies a function to every element of an iterable (such as a list).

#Syntax
#map(function, iterable)
#function → the function to apply
#iterable → list, tuple, etc.

#Since map() returns a map object, you usually convert it into a list.
numbers = [1, 2, 3, 4, 5]

result = map(lambda element: element**2, numbers)

print(list(result))
```

```python
#filter()
#Purpose

#filter() selects only those elements that satisfy a condition.

#Syntax
#filter(function, iterable)

#The function must return:

#True → keep the element
#False → discard the element

#Example 1: Keep only even numbers
numbers = [1,2,3,4,5,6,7,8]

even = filter(lambda x: x % 2 == 0, numbers)

print(list(even))
```

```python
#Example 2: Find students who passed

marks = [45, 82, 60, 35, 90]

passed = filter(lambda x: x >= 50, marks)

print(list(passed))
```

```python
# reduce()
#Purpose

#reduce() combines all elements into a single value.

#Unlike map() and filter(), reduce() is part of the functools module.

#Import it
from functools import reduce

#Syntax
#reduce(function, iterable)
#The function takes two arguments:
lambda x, y: ...
#where:

#x = accumulated result
#y = next element

#Example 1: Sum all numbers
from functools import reduce

numbers = [1,2,3,4,5]

total = reduce(lambda x, y: x + y, numbers)

print(total)
```
